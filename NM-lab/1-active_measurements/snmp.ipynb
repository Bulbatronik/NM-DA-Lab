This is a quick example on how using SNMP. We will interrogate a (simulated) Cisco 7513 router hosted at snmp.live.gambitcommunications.com

First we'll need to install snmp and the MIB useful for retrieving measurements 

```python
!apt-get install snmp
!apt-get install snmp-mibs-downloader
```

Now we can start using SNMP. We will use the command snmpwalk, which allows to retrieve all possible information from the device.

```python
!snmpwalk -c public -v1 snmp.live.gambitcommunications.com
```

As you see, 'walking' may be a long operation... Luckily we can use the MIB for directly quering some OIDs

As an example the OID 1.3.6.1.2.1.2.1.0 refers to the total number of network interfaces in the router. It can also be queried with the name IF-MIB::ifNumber (more info here: http://www.net-snmp.org/docs/mibs/interfaces.html)

```python
!snmpwalk -c public -v1 snmp.live.gambitcommunications.com IF-MIB::ifNumber
```

Now we can try to get information for any of the network interfaces... say the 10th interface

```python
!snmpwalk -c public -v1 snmp.live.gambitcommunications.com IF-MIB::ifDescr.10
!snmpwalk -c public -v1 snmp.live.gambitcommunications.com IF-MIB::ifType.10
!snmpwalk -c public -v1 snmp.live.gambitcommunications.com IF-MIB::ifSpeed.10
!snmpwalk -c public -v1 snmp.live.gambitcommunications.com IF-MIB::ifInOctets.10
!snmpwalk -c public -v1 snmp.live.gambitcommunications.com IF-MIB::ifInErrors.10
```
