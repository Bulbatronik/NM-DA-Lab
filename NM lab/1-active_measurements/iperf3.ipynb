This notebook shows how to perform active measurements with iperf3. Unfortunately, there is no python native version of iperf, so we will use the command line version (Colab allows to do this as we shall see...)

First we will install iperf3 in the Virtual Machine system:

```python
!apt-get install iperf3
```

Now we are ready to use iperf3. We will use it in client mode (specifying the -c parameter) since the Colab VM is not reachable from the external (afaik). 

We now need a public, reachable iperf3 server to perform some measurements again. Searching on the internet I found iperf3 active servers [at this page](https://iperf.fr/iperf-servers.php)

(n.b. each server accepts only one connection at a time so it might not work right away)

First we'll try to perform a TCP bandwidth test to one server (pay attention to the listening port in the web page!):



```python
!iperf3 -c paris.testdebit.info -p 9200 --get-server-output
```

Good! Now we'll perform a UDP bandwidth test (with the -u parameter). We'll send 1000bytes datagrams (-l 1000) at a specific rate of 10 Mbps (-b 10000000)...

```python
!iperf3 -c paris.testdebit.info -p 9200 -u --get-server-output -l 1000 -b 10000000
```
